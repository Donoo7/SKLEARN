```python
import pandas as pd
df = pd.DataFrame({
    "name": ["Swift", "Verna", "Polo", "Creta", "Innova"],
    "brand": ["Suzuki", "Hyundai", "VW", "Hyundai", "Toyota"],
    "color": ["Red", "White", "Red", "Black", "White"]
})
df
```

```python
from sklearn.preprocessing import OneHotEncoder

df_onehotencode = pd.get_dummies(data = df, columns = ["brand", "color"],prefix = "is")
df_onehotencode
```

```python
col_to_encode = ["brand", "color"]
new_encode_col_names = []

for col in col_to_encode:
    new_encode_col_names += [f"is_{category}" for category in df[col].unique().tolist()]

print(new_encode_col_names)
['is_Suzuki', 'is_Hyundai', 'is_VW', 'is_Toyota', 'is_Red', 'is_White', 'is_Black']
one_hot_encode = OneHotEncoder(sparse = False, handle_unknown = 'ignore')
encode_col = one_hot_encode.fit_transform(df[col_to_encode])

df_encoded = pd.DataFrame(encode_col, columns = new_encode_col_names)
df_one_hot_encode = df.join(df_encoded)
print(df_one_hot_encode)
```
